<h1>Lab 1 Deep Learing for Computer Vision<h1>

```python
import matplotlib.pyplot as plt
import random
```

```python
my_list = [6, 3, 2,3, 6, 11, 18]

plt.plot(my_list)
plt.title("Y data")
plt.xlabel('Time')
plt.ylabel('Y')
```

<h2>len()<h2>

```python
# print range of list

print('my_list data {}'.format(my_list))
print('len of my_list data {}'.format(len(my_list)))
```

<h2>append()<h2>

```python
# Append data to list

for i in range(5):
    my_list.append(random.randint(1,20))
    
print("List after append:", my_list)

plt.plot(my_list, color='red')
plt.title("Y data")
plt.xlabel('Time')
plt.ylabel('Y')
```

<h2>insert()<h2>

```python
# Insert 5 data to 3 in place of list

for i in range(5):
    my_list.insert(3,random.randint(1,20))
    
print("List after insert:", my_list)
plt.plot(my_list, color='red')
plt.title("Y data")
plt.xlabel('Time')
plt.ylabel('Y')
```

<h2>extend()<h2>

```python
# Extend 30, 40, 50 to list

elements_to_add = [30, 40, 50]

my_list.extend(elements_to_add)

print("List after extend:", my_list)
plt.plot(my_list, color='red')
plt.title("Y data")
plt.xlabel('Time')
plt.ylabel('Y')
```

<h2>pop()<h2>

```python
# Pop 3, 5, 7, 9 from list

indices_to_remove = [3, 5, 7, 9]
for index in indices_to_remove:
    if index < len(my_list):
        my_list.pop(index)
        print(f'Removed element at index {index}')
    
print("List after pop:", my_list)
plt.plot(my_list, color='red')
plt.title("Y data")
plt.xlabel('Time')
plt.ylabel('Y')
```

<h2>remove()<h2>

```python
# Remove 30, 40 ,50 from list

values_to_remove = [30, 40, 50]

for value in values_to_remove:
    if value in my_list:
        my_list.remove(value)

print("List after removing values:", my_list)
plt.plot(my_list, color='red')
plt.title("Y data")
plt.xlabel('Time')
plt.ylabel('Y')
```

<h2>sorted()<h2>

```python
# Sorted data in list from min to max

sorted_list = sorted(my_list)

print("Original List:", my_list)
print("Sorted List:", sorted_list)

print("List after Sorted values:", sorted_list)
plt.plot(sorted_list, color='red')
plt.title("Y data")
plt.xlabel('Time')
plt.ylabel('Y')
```

<h2>sort()<h2>

```python
# Sort data in list from min to max

my_list.sort()

print("Sorted List (in-place):", my_list)
plt.plot(my_list, color='red')
plt.title("Y data")
plt.xlabel('Time')
plt.ylabel('Y')
```

<h2>sort() reverse<h2>

```python
# Invert sort data in list from min to max invert to max to min

my_list.sort(reverse=True)

print("Sorted List (invert):", my_list)
plt.plot(my_list, color='red')
plt.title("Y data")
plt.xlabel('Time')
plt.ylabel('Y')
```

<h2>reverse()<h2>

```python
# Reverse the list in-place
my_list.reverse()

print("Reversed List (in-place):", my_list)
plt.plot(my_list, color='red')
plt.title("Y data")
plt.xlabel('Time')
plt.ylabel('Y')
```

<h2>index()<h2>

```python
# Find the index of the first occurrence of the value 4
index = my_list.index(4)

# Print the index
print("Data of my list : {}".format(my_list))
print("Index of the first occurrence of 4:", index)
```

<h2>count()<h2>

```python
# Count the number of occurrences of the value 4
count = my_list.count(4)

# Print the conut
print("Data of my list : {}".format(my_list))
print("Count of 4 in the list:", count)
```

<h2>copy()<h2>

```python
# Coppy orginal list
copied_list = my_list.copy()


print("Original List:", my_list)
print("Copied List:", copied_list)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,5))

ax1.plot(my_list, color='blue')
ax1.set_title("Original List")
ax1.set_xlabel('Time')
ax1.set_ylabel('Y')

ax2.plot(copied_list, color='red')
ax2.set_title("Copied List")
ax2.set_xlabel('Time')
ax2.set_ylabel('Y')
```

<h2>sum()<h2>

```python
# Calculate the sum of all elements in the list
total_sum = sum(my_list)

print("Sum of the elements:", total_sum)
```

<h2>min()<h2>

```python
# Find the minimum value in the list
min_value = min(my_list)

print("Minimum value:", min_value)
```

<h2>max()<h2>

```python
# Find the maximum value in the list
max_value = max(my_list)

print("Maximum value:", max_value)
```
